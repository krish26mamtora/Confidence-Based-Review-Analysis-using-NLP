```python
import os
import pandas as pd

subfolders = [f.name for f in os.scandir(os.getcwd()) if f.is_dir()]
folder_mapping = []

for folder_name in subfolders:
    v1_exists = os.path.exists(os.path.join(os.getcwd(), folder_name, 'v1'))

    # Check for reviews.json and meta.json within v1
    reviews_exists = False
    meta_exists = False
    if v1_exists:  # Only check if v1 exists
        reviews_exists = os.path.exists(os.path.join(os.getcwd(), folder_name, 'v1', 'reviews.json'))
        meta_exists = os.path.exists(os.path.join(os.getcwd(), folder_name, 'v1', 'meta.json'))

    folder_mapping.append([folder_name, v1_exists, reviews_exists, meta_exists])

df = pd.DataFrame(folder_mapping, columns=['Folder Name', 'Contains v1', 'Contains reviews.json', 'Contains meta.json'])
display(df)
```

```python
import pandas as pd
import json
from google.colab import files
```

```python
df['abstract'] = ''
df['title'] = ''

for index, row in df.iterrows():
    if row['Contains v1'] and row['Contains meta.json']:
        meta_json_path = os.path.join(os.getcwd(), row['Folder Name'], 'v1', 'meta.json')

        try:
            with open(meta_json_path, 'r') as meta_file:
                meta_data = json.load(meta_file)
                df.loc[index, 'abstract'] = meta_data.get('abstract', '')  # Handle missing key
                df.loc[index, 'title'] = meta_data.get('title', '')      # Handle missing key
        except FileNotFoundError:
            print(f"meta.json not found in {row['Folder Name']}/v1")
        except json.JSONDecodeError:
            print(f"Error decoding JSON in {row['Folder Name']}/v1/meta.json")


```

```python
import pandas as pd
import json
import os

# Get the list of subfolders in the current working directory
subfolders = [f.name for f in os.scandir(os.getcwd()) if f.is_dir()]

# Initialize an empty list to store the extracted data
extracted_data = []

# Iterate through each subfolder
for folder_name in subfolders:
    v1_folder_path = os.path.join(os.getcwd(), folder_name, 'v1')
    reviews_json_path = os.path.join(v1_folder_path, 'reviews.json')

    if os.path.exists(reviews_json_path):
        try:
            with open(reviews_json_path, 'r') as f:
                reviews_data = json.load(f)
                for key, review in reviews_data.items():
                    # Extract the required fields, handle potential missing keys gracefully
                    rid = review.get('rid', '')
                    paper_summary = review['report'].get('paper_summary', '')
                    summary_of_strengths = review['report'].get('summary_of_strengths', '')
                    summary_of_weaknesses = review['report'].get('summary_of_weaknesses', '')
                    confidence = review['meta'].get('confidence', '')
                    overall_assessment = review['scores'].get('overall_assessment', '')

                    # Append the extracted data along with the folder name
                    extracted_data.append([
                        folder_name,  # subfolder name
                        rid, paper_summary, summary_of_strengths,
                        summary_of_weaknesses, overall_assessment , confidence
                    ])
        except json.JSONDecodeError:
            print(f"Error decoding JSON in {reviews_json_path}")
        except KeyError as e:
            print(f"KeyError in {reviews_json_path}: {e}")  # Print the specific missing key

# Create a Pandas DataFrame
df_reviews = pd.DataFrame(extracted_data, columns=[
    'subfolder', 'rid', 'paper_summary', 'summary_of_strengths',
    'summary_of_weaknesses', 'overall_assessment', 'confidence'
])
```

```python
from openpyxl.cell.cell import ILLEGAL_CHARACTERS_RE

# Clean the DataFrame
df_reviews_clean = df_reviews.applymap(lambda x: ILLEGAL_CHARACTERS_RE.sub('', x) if isinstance(x, str) else x)

# Now write to Excel
df_reviews_clean.to_excel('reviews.xlsx', index=False)
```
