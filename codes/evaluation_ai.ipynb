```python
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Step 1: Load the Excel file
# Replace 'your_file.xlsx' with your actual Excel file name
df = pd.read_excel("/content/reviews_with_bert_jaccard_scibert (1).xlsx")

# Step 2: Define the true column and predicted columns
true_col = 'confidence'
predicted_cols = ['bert_confidence', 'jaccard_confidence', 'scibert_confidence']

# Step 3: Compute MSE and MAE for each model
for col in predicted_cols:
    mse = mean_squared_error(df[true_col], df[col])
    mae = mean_absolute_error(df[true_col], df[col])
    print(f"\nMetrics for {col}:")
    print(f"  MSE: {mse:.4f}")
    print(f"  MAE: {mae:.4f}")

```
