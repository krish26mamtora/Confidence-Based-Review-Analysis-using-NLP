```python
import matplotlib.pyplot as plt

# Correlation values
models = ['BERT', 'Jaccard', 'SciBERT']
correlations = [-0.0376, -0.0990, -0.0523]

# Plotting
plt.figure(figsize=(8, 5))
bars = plt.bar(models, correlations, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.axhline(0, color='gray', linestyle='--')

# Annotate bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., 1.05*height,
             f'{height:.3f}', ha='center', va='bottom')

plt.title('Correlation between Model Output and Confidence Scores')
plt.ylabel('Pearson Correlation')
plt.xlabel('Models')
plt.ylim(-0.12, 0.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
